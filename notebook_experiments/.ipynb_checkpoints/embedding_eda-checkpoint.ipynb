```python
import pandas as pd 
import numpy as np 
import gensim 
import spacy 
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
## NN based spanish sentiment scorer built on Tensorflow + Keras
from sentiment_analysis_spanish import sentiment_analysis
from pre_process.text_utils import *

nlp_spanish = spacy.load("es_core_news_sm")

%load_ext autoreload
%autoreload 2
```

## Use La Jornada web scraped articles on gender violence which contains around ~2000 articles

```python
jornada_df = pd.read_excel("../data/Mexico/la_jornada_scraped.xlsx")
universal_df = pd.read_excel("../data/Mexico/el_universal_scraped.xlsx")
```

```python
jornada_df.head(5)
```

```python
# Join both text columns
jornada_df["title"] = jornada_df["title"].str.pad(1, side ='right', fillchar =' ')
jornada_df["complete_text"]  = jornada_df.title.str.cat(jornada_df.text).astype(str)

universal_df["complete_text"]  = universal_df.title.str.cat(universal_df.text).astype(str)
```

```python
# Normalize and tokenize all documents
jornada_df['tokenized_sentences'] = jornada_df['complete_text'].apply(lambda x: 
                                                                      [word_tokenize(s, model=nlp_spanish)
                                                                      for s in sent_tokenize(x)])
jornada_df['normalized_sentences'] = jornada_df['tokenized_sentences'].apply(lambda x:
                                                                            [normalize_tokens(s, model=nlp_spanish)
                                                                            for s in x])
```

```python
universal_df['tokenized_sentences'] = universal_df['complete_text'].apply(lambda x: 
                                                                      [word_tokenize(s, model=nlp_spanish)
                                                                      for s in sent_tokenize(x)])
universal_df['normalized_sentences'] = universal_df['tokenized_sentences'].apply(lambda x:
                                                                            [normalize_tokens(s, model=nlp_spanish)
                                                                            for s in x])
```

```python
laJornada_w2v  = gensim.models.word2vec.Word2Vec(jornada_df['normalized_sentences'].sum(),window=50, workers=4)
elUniversal_w2v  = gensim.models.word2vec.Word2Vec(universal_df['normalized_sentences'].sum(),window=50, workers=4)
```

```python
# Initialize sentiment scorer object
SentimentScorer = sentiment_analysis.SentimentAnalysisSpanish()
```

```python
laJornada_w2v.wv.most_similar("hombre")
```

```python
elUniversal_w2v.wv.most_similar("hombre")
```

```python
words_of_interest = ["hombre", "mujer", "feminicidio", "abuso"]

for word in words_of_interest:
    print()
    print(f"Finding words most similar to '{word}' according to its embedding")
    try:
        print("...................La Jornada.................................")
        words_tupple = laJornada_w2v.wv.most_similar(word)
        rep_words = [word for word, score in words_tupple]
        print(words_tupple)
        sentiment = SentimentScorer.sentiment(" ".join(rep_words))
        print(f"Sentiment {sentiment}")
        print("...................El Universal................................")
        words_tupple = elUniversal_w2v.wv.most_similar(word)
        rep_words = [word for word, score in words_tupple]
        print(words_tupple)
        sentiment = SentimentScorer.sentiment(" ".join(rep_words))
        print(f"Sentiment {sentiment}")
        print()

    except:
        continue
```

## Visualizing Word Embeddings using dimensionality reduction

```python
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Retrieve vectors
X_jornada = laJornada_w2v[laJornada_w2v.wv.vocab]
X_universal = elUniversal_w2v[elUniversal_w2v.wv.vocab]
```

```python
pca = PCA(n_components=2)
result_la_jornada = pca.fit_transform(X_jornada)
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x=result_la_jornada[:, 0], y=result_la_jornada[:, 1])
for i, word in enumerate(words_of_interest):
    plt.scatter(result_la_jornada[i, 0], result_la_jornada[i, 1], marker='o', s=100)
    plt.annotate(word, xy=(result_la_jornada[i, 0], result_la_jornada[i, 1]))
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("La Jornada Word Embeddings")
```

```python
result_el_universal = pca.fit_transform(X_universal)
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x=result_el_universal[:, 0], y=result_el_universal[:, 1])
for i, word in enumerate(words_of_interest):
    plt.scatter(result_el_universal[i, 0], result_el_universal[i, 1], marker='o', s=100)
    plt.annotate(word, xy=(result_el_universal[i, 0], result_el_universal[i, 1]))
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("El Universal Word Embeddings")
plt.xlim([-2, 6])
plt.ylim([-8, 8])
```
